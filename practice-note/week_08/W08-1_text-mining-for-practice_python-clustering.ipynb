```python
from teanaps.nlp.SyntaxAnalyzer import SyntaxAnalyzer

sa = SyntaxAnalyzer()

document_list = []
tokenized_sentence_list = []

POS_LIST = ["NNG", "NNP"]

f = open("teanaps/data/article_sample.txt", encoding="utf-8")

for line in f:
    line = line.strip()
    col = line.split("\t")
    label = col[0]
    source = col[1]
    datetime = col[2]
    title = col[3]
    content = col[4]
    document_list.append([label, source, datetime, title, content])
    tagged_word_list = sa.parse(content)
    tokenized_sentence = ""
    for word, pos in tagged_word_list:
        if pos in POS_LIST:
            tokenized_sentence += " " + word
    tokenized_sentence_list.append(tokenized_sentence.strip())
f.close()

tokenized_sentence_list[:5]
```

```python
import pandas as pd

df_article = pd.DataFrame(document_list, columns = ["label", "source", "datetime", "title", "content"])
df_article[:5]
```

```python
from teanaps.text_analysis.DocumentClustering import DocumentClustering

dc = DocumentClustering()

NUM_CLUSTERS = 3
NUM_INIT = 10
MAX_ITERATIONS = 300

result = dc.kmeans_clustering(tokenized_sentence_list, NUM_CLUSTERS, NUM_INIT, MAX_ITERATIONS)
result
```

```python
MAX_CLUSTERS = 10
NUM_INIT = 10
MAX_ITERATIONS = 300

inertia_list = dc.inertia_transition(tokenized_sentence_list, MAX_CLUSTERS, NUM_INIT, MAX_ITERATIONS)
inertia_list
```

```python
dc.set_plotly()
dc.get_inertia_transition_graph(inertia_list)
```

```python
PREDICT_LIST = result["predict_list"]
DF_DOCUMENT_LIST = df_article

df_result = dc.get_tfidf_tsne(tokenized_sentence_list, PREDICT_LIST, DF_DOCUMENT_LIST)
df_result
```

```python
dc.set_plotly()
dc.get_kmeans_graph(df_result, "predict")
```

```python
dc.set_plotly()
dc.get_kmeans_graph(df_result, "label")
```
