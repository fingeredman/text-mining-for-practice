- Python으로 문서에서 단어의 동시출현빈도를 계산하고 네트워크를 생성하는 방법에 대해 다룹니다.

---

```python
from teanaps.nlp import SyntaxAnalyzer

sa = SyntaxAnalyzer()

tokenized_sentence_list = []

PATH = "teanaps/data/article_sample.txt"
POS_LIST = ["NNG", "NNP"]

f = open(PATH, encoding="utf-8")

for line in f:
    line = line.strip()
    col = line.split("\t")
    label = col[0]
    source = col[1]
    datetime = col[2]
    title = col[3]
    content = col[4]
    tagged_word_list = sa.parse(content)
    tokenized_sentence = ""
    for word, pos, loc in tagged_word_list:
        if pos in POS_LIST:
            tokenized_sentence += " " + word
    tokenized_sentence_list.append(tokenized_sentence.strip())
f.close()

tokenized_sentence_list[:5]
```

```python
from teanaps.text_analysis import CoWordCalculator

co = CoWordCalculator()
```

```python
co.calculation_co_matrix(tokenized_sentence_list)[:20]
```
