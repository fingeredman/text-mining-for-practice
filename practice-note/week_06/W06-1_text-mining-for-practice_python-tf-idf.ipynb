### 1. 단어빈도와 TF-IDF 계산하기

---

```python
# TEANAPS 라이브러리를 불러옵니다.
from teanaps.nlp import MorphologicalAnalyzer
from teanaps.nlp import Processing

pro = Processing()
ma = MorphologicalAnalyzer()
#ma.set_tagger("okt")

tokenized_sentence_list = []
PATH = "data/article_sample.txt"
POS_LIST = ["NNG", "NNP"]

f = open(PATH, encoding="utf-8")

# 파일에서 기사본문만 추출하여 명사형 문장을 생성합니다.
for line in f:
    line = line.strip()
    col = line.split("\t") # [label, souce, datetime, title, content]
    label = col[0]         # 기사 검색어
    source = col[1]        # 언론사
    datetime = col[2]      # 작성일자
    title = col[3]         # 제목
    content = col[4]       # 본문
    tagged_word_list = ma.parse(content)
    tokenized_sentence = pro.get_plain_text(tagged_word_list, pos_list=POS_LIST, tag=False)
    tokenized_sentence_list.append(tokenized_sentence)

f.close()

tokenized_sentence_list[:3]
```

```python
from teanaps.text_analysis import TfidfCalculator

tfidf = TfidfCalculator()
```

```python
# TF/TF-IDF 값을 계산합니다.
tfidf.calculation_tfidf(tokenized_sentence_list, tfidf_count=100, tfidf_state=True)
```

```python
# 단어 x TF 매트릭스를 생성합니다.
tfidf.get_tf_matrix()
```

```python
# 단어 x TF-IDF 매트릭스를 생성합니다.
tfidf.get_tfidf_matrix()
```

```python
# TF 계산결과를 딕셔너리 형태로 불러옵니다.
tfidf.get_tf_dict()
```

```python
# TF 계산결과를 리스트 형태로 불러옵니다.
tfidf.get_tf_list()[:20]
```

```python
# TF-IDF 계산결과를 딕셔너리 형태로 불러옵니다.
tfidf.get_tfidf_dict()
```

```python
# TF-IDF 계산결과를 리스트 형태로 불러옵니다.
tfidf.get_tfidf_list()[:20]
```

```python
# 전체 문서의 단어주머니를 불러옵니다.
tfidf.get_word_list()[:20]
```

```python
# TF/TF-IDF 계산결과를 그래프로 표현합니다.
# 그래프에 표현할 빈도기준 Top N 개를 숫자로 입력합니다.
tfidf.draw_tfidf(100)
```

```python
# TF-IDF 기준 워드클라우드를 출력합니다.
count_dict = tfidf.get_tfidf_dict()
tfidf.get_wordcloud(count_dict)
```

```python
# TF 기준 워드클라우드를 출력합니다.
count_dict = tfidf.get_tf_dict()
tfidf.get_wordcloud(count_dict)
```

### 2. 불용어 처리하기

---

```python
from teanaps.nlp import Processing

pro = Processing()
```

```python
# 전체 불용어 리스트를 불러옵니다.
result = pro.get_stopword()
print(result[-10:])
```

```python
# 불용어를 추가합니다.
pro.add_stopword("가")

result = pro.get_stopword()
print(result[-10:])
```

```python
# 불용어를 리스트 단위로 추가합니다.
pro.add_stopword(["으로", "로서", "때문에"])

result = pro.get_stopword()
print(result[-10:])
```

```python
# 불용어를 제거합니다.
pro.remove_stopword("때문에")

result = pro.get_stopword()
print(result[-10:])
```

```python
# 불용어를 리스트 단위로 제거합니다.
pro.remove_stopword(["은", "는", "이", "가"])

result = pro.get_stopword()
print(result[-10:])
```

```python
# 전체 불용어 리스트를 삭제합니다.
pro.clear_stopword()

result = pro.get_stopword()
print(result[-10:])
```

```python
# 불용어 리스트를 초기화합니다.
pro.set_org_stopword()

result = pro.get_stopword()
print(result[-10:])
```

```python
# 불용어 여부를 확인합니다.
result = pro.is_stopword("은")
print(result)
```

```python
result = pro.is_stopword("없는단어")
print(result)
```
