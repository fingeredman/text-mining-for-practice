### 1. 뉴스기사 불러오기

---

```python
from teanaps.handler import FileHandler

fh = FileHandler()

PATH = "data/article_sample.txt"
line_list = fh.load_txt(PATH)
```

```python
line_list[:3]
```

### 2. 뉴스기사 요약하기

---

```python
from teanaps.text_analysis import DocumentSummarizer

ds = DocumentSummarizer()
```

```python
article = line_list[50]
article
```

```python
content = article[4]
content
```

#### 2.1. 요약하기: TextRank

---

```python
SUMMARIZER_TYPE = "textrank"
MAX_SENTENCES = 3

result = ds.summarize(SUMMARIZER_TYPE, MAX_SENTENCES, document=content)
result
```

#### 2.2. 요약하기: LSA

---

```python
SUMMARIZER_TYPE = "lsa"
MAX_SENTENCES = 3

result = ds.summarize(SUMMARIZER_TYPE, MAX_SENTENCES, document=content)
result
```
